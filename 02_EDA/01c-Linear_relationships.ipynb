# Visualization of a linear relationship

Working on *IBM_HR_ATTRITION.csv*.

1. Import usual librairies

```python
import pandas as pd
pd.set_option('display.max_columns', 500)

import numpy as np

import seaborn as sns
sns.set_theme(style="whitegrid")

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("src\IBM_HR_ATTRITION.csv")
df
```

```python
# def yes_no(x):
#     if x == "Yes":
#         return 1
#     else:
#         if x =="No":
#             return 0
        
# df["Attrition"] = df["Attrition"].apply(lambda x: yes_no(x))
```

```python
df["Attrition"] = df["Attrition"].astype(str).replace({"Yes": 1, "No": 0})
df
```

2. We would like to know the probability of leaving the business in relation to the distance from home. Use _lmplot()_ to view this. What can you conclude?

```python
sns.lmplot(data= df, y= "Attrition", x= "DistanceFromHome", aspect=30/10)
plt.show()
```

```python
sns.lmplot(data= df, x= "DistanceFromHome", y= "Attrition", logistic= True, aspect=30/10)
plt.show()
```

```python
sns.lmplot(data= df, x= "DistanceFromHome", y= "Attrition", logistic= True, scatter_kws={"alpha":.05}, aspect=30/10)
plt.show()
```

3. Try it this time with the number of years spent with the Manager (_YearsWithCurrManager)_, what can you conclude?

```python
sns.lmplot(data= df, x= "YearsWithCurrManager", y= "Attrition", logistic= True, scatter_kws={"alpha":.05}, aspect=30/10)
plt.show()
```
