````python
rapport_1 = df["ACTEU"].value_counts(dropna= False).iloc[2] / df["ACTEU"].value_counts(dropna= False).iloc[1]
rapport_2 = df["ACTEU"].value_counts(dropna= False).iloc[2] / ( df["ACTEU"].value_counts(dropna= False).iloc[1] + df["ACTEU"].value_counts(dropna= False).iloc[0])

print("['Chômeurs / Actifs Occupés', 'Chômeurs / Actifs + Inactifs']")
print(f"Rapport {[round(rapport_1, 2), round(rapport_2, 2)]}")
````

* Let's end this exercise by looking at the distribution of desired hours for those looking for part-time work. To do this, use the HREC column. 

  * First try to simply create a histogram without the density function. 
  * To create the density function, you will need to use the following formula: 
      ```y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))```
     
     where *mu* is the mean of the distribution and *sigma* is the standard deviation.  You can help yourself with this tutorial [histogram with Density](https://matplotlib.org/3.1.1/gallery/statistics/histogram_features.html)

````python
df["HREC"].value_counts(dropna= True).sort_index(ascending= True)
````

````python
plt.figure(figsize=(15,5))

data = df["HREC"]
mu = data.mean()
sigma = data.std()
nbins = 20

plt.hist(x= data, bins= nbins, density= True)

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (nbins - mu))**2))
plt.plot(nbins, y, "--")

plt.title(f"Histogram: $\mu={round(mu, 2)}$, $\sigma={round(sigma, 2)}$",loc= "center", pad= 15, weight="bold")
plt.xlabel("")
plt.ylabel("")
plt.grid(None)

plt.show()
````
